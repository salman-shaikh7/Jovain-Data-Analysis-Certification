## Function 1 - np.stack
used to join the sequence of an array along a new axis

```python
'''
#AXIS =0
array([[[ 1.,  2.],
        [ 3.,  4.]],

       [[ 5., 11.],
        [ 8.,  9.]]])
'''
arr1 = [
        [1, 2], 
        [3, 4]
        ]

arr2 = [
        [5, 11], 
        [8, 9]
        ]

np.stack((arr1, arr2),axis=1)
```

Explanation about example2: Now we will try to use stack on 2 array and we will axis eqaul to 0 axis=0

```python
# Example 2 - working
arr3=[
      [1,2,3],
      [4,5,6]
     ]
arr4=[
      [7,8,9],
      [10,11,12]
     ]
```

```python
np.stack((arr3+arr4),axis=0)
```

Explanation about example3: It will throw error "all input arrays must have the same shape" 

```python
# Example 3 - breaking (to illustrate when it breaks)
arr1 = [[1, 2], 
        [3, 4.]]

arr2 = [[5, 6, 7], 
        [8, 9, 10]]

np.stack((arr1, arr2), axis=0)
```
