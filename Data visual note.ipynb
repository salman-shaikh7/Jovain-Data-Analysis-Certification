```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
arr1=[1,2,1,4,8,6,7,10]
arr2=['a','b','c','d','e','f','g','h']
```

```python
plt.plot(arr2,arr1);
plt.xlabel("LETTER")
plt.title("LETTER REPETATION")
plt.ylabel("FREQ");
```

```python
arr1=[1,2,1,4,8,6,7,10]
arr1_5=[5,6,1,7,7,1,7,2]
arr2=['a','b','c','d','e','f','g','h']
```

```python
plt.plot(arr2,arr1);
plt.plot(arr2,arr1_5);
plt.xlabel("LETTER")
plt.title("LETTER REPETATION")
plt.ylabel("FREQ");

plt.legend(['arr1', 'arr1-5']);
```

```python
#plt.plot(arr2,arr1,marker='o',linestyle='--',color='green');
plt.figure(figsize=(6,4));

plt.plot(arr2,arr1,'o--y');
plt.plot(arr2,arr1_5,'>-b');

plt.xlabel("LETTER")
plt.ylabel("FREQ");

plt.title("LETTER REPETATION")

plt.legend(['arr1', 'arr1-5']);
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(6,4));
plt.plot(arr2,arr1,'oy');#o marker and yellow color
```
